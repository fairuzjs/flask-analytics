16. Tentukan kabupaten/kota dengan total `jumlah_penderita_dm` tertinggi dan terendah di seluruh tahun.

```python
import numpy as np

conditions = [
    df['jumlah_penderita_dm'] < 50000,
    (df['jumlah_penderita_dm'] >= 50000) & (df['jumlah_penderita_dm'] < 100000),
    df['jumlah_penderita_dm'] >= 100000
]
choices = ['Rendah', 'Sedang', 'Tinggi']

df['kategori_dm'] = np.select(conditions, choices, default='Tidak Dikategorikan')
display(df[['nama_kabupaten_kota', 'jumlah_penderita_dm', 'kategori_dm']].head())
```

- `< 50.000`  "Rendah"
- `50.00099.999`  "Sedang"
- `>= 100.000`  "Tinggi"

```python
total_dm_by_kab_kota = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum()

kab_kota_tertinggi = total_dm_by_kab_kota.idxmax()
jumlah_tertinggi = total_dm_by_kab_kota.max()

kab_kota_terendah = total_dm_by_kab_kota.idxmin()
jumlah_terendah = total_dm_by_kab_kota.min()

print(f"Kabupaten/Kota dengan total penderita DM tertinggi: {kab_kota_tertinggi} ({jumlah_tertinggi} orang)")
print(f"Kabupaten/Kota dengan total penderita DM terendah: {kab_kota_terendah} ({jumlah_terendah} orang)")
```

18. Untuk setiap tahun, hitung persentase jumlah penderita DM per kabupaten/kota terhadap total jumlah penderita DM di tahun tersebut, lalu simpan dalam kolom baru bernama `persentase_tahun`.

```python
total_dm_per_year = df.groupby('tahun')['jumlah_penderita_dm'].transform('sum')
df['persentase_tahun'] = (df['jumlah_penderita_dm'] / total_dm_per_year * 100).round(2)
display(df[['tahun', 'nama_kabupaten_kota', 'jumlah_penderita_dm', 'persentase_tahun']].head())
```

```python
import pandas as pd
import numpy as np
import json

# ==========================================
# 1. PERSIAPAN DATA
# ==========================================
# Ganti nama file ini sesuai dengan file CSV yang Anda upload di Colab
filename_csv = '/content/dinkes-od_17448_jml_penderita_diabetes_melitus_brdsrkn_kabupatenko_v2_data.csv'

try:
    # Membaca data (Soal 1)
    df = pd.read_csv(filename_csv)
    print(" Data berhasil dimuat!")
except FileNotFoundError:
    print(f" Error: File '{filename_csv}' tidak ditemukan. Pastikan sudah upload file CSV.")
    # Membuat dummy data untuk contoh agar kode tidak error saat dijalankan tanpa file
    data_dummy = {
        'nama_kabupaten_kota': ['Kab. A', 'Kab. B', 'Kab. C'] * 6,
        'jumlah_penderita_dm': np.random.randint(1000, 50000, 18),
        'tahun': [2019, 2020, 2021, 2022, 2023, 2024] * 3,
        'kategori_dm': ['Rendah', 'Sedang', 'Tinggi'] * 6
    }
    df = pd.DataFrame(data_dummy)

# ==========================================
# 2. ANALISIS DATA (SOAL 1 - 19)
# ==========================================

# --- A. Inspeksi Dasar (Soal 1-5) ---
# 1. Bentuk Data
total_baris, total_kolom = df.shape
# 2. Info Kolom
daftar_kolom = df.columns.tolist()
# 3. Wilayah Unik
wilayah_unik = df['nama_kabupaten_kota'].nunique()
daftar_wilayah = df['nama_kabupaten_kota'].unique().tolist()
# 4. Cek Null
total_null = int(df.isnull().sum().sum())
# 5. Cek Duplikat
total_duplikat = int(df.duplicated().sum())

# --- B. Filter & Fokus Data (Soal 6-7) ---
# Fokus pada data tahun terbaru (misal 2024) untuk analisis status terkini
tahun_terbaru = df['tahun'].max()
df_terbaru = df[df['tahun'] == tahun_terbaru].copy()

# --- C. Nilai Ekstrem (Soal 8-9) ---
# 8. Nilai Tertinggi (Max)
row_max = df_terbaru.loc[df_terbaru['jumlah_penderita_dm'].idxmax()]
max_nama = row_max['nama_kabupaten_kota']
max_nilai = int(row_max['jumlah_penderita_dm'])

# 9. Nilai Terendah (Min)
row_min = df_terbaru.loc[df_terbaru['jumlah_penderita_dm'].idxmin()]
min_nama = row_min['nama_kabupaten_kota']
min_nilai = int(row_min['jumlah_penderita_dm'])

# --- D. Statistik Pusat (Soal 10-12) ---
# 10. Mean (Rata-rata)
mean_val = df_terbaru['jumlah_penderita_dm'].mean()
# 11. Median (Nilai Tengah)
median_val = df_terbaru['jumlah_penderita_dm'].median()
# 12. Std Dev (Simpangan Baku)
std_val = df_terbaru['jumlah_penderita_dm'].std()

# --- E. Sebaran Data / Kuantil (Soal 13-14) ---
# 13. Q1 (25%)
q1_val = df_terbaru['jumlah_penderita_dm'].quantile(0.25)
# 14. Q3 (75%)
q3_val = df_terbaru['jumlah_penderita_dm'].quantile(0.75)
# IQR
iqr_val = q3_val - q1_val

# --- F. Distribusi Kategori (Soal 15) ---
# Asumsi kolom 'kategori_dm' sudah ada. Jika belum, kita buat logic sederhana.
if 'kategori_dm' not in df.columns:
    # Logic pembuatan kategori jika belum ada (berdasarkan kuartil)
    def get_kategori(x):
        if x > q3_val: return 'Tinggi'
        elif x < q1_val: return 'Rendah'
        else: return 'Sedang'
    df_terbaru['kategori_dm'] = df_terbaru['jumlah_penderita_dm'].apply(get_kategori)

kategori_counts = df_terbaru['kategori_dm'].value_counts()
distribusi_list = []
colors = {'Tinggi': 'bg-red-500', 'Sedang': 'bg-orange-500', 'Rendah': 'bg-green-500'}

for kat, jum in kategori_counts.items():
    distribusi_list.append({
        "label": kat,
        "jumlah": int(jum),
        "warna": colors.get(kat, 'bg-gray-400')
    })

# --- G. Agregasi & Pola (Soal 16-19) ---
# 16. Total Kasus per Tahun (Tren)
tren_tahun = df.groupby('tahun')['jumlah_penderita_dm'].sum().reset_index()
tren_tahun_dict = tren_tahun.to_dict(orient='records')

# 17. Total Kasus Seluruh Wilayah (Kumulatif 2019-2024)
total_kasus_jabar = int(df['jumlah_penderita_dm'].sum())

# 18. Top 5 Wilayah Kumulatif Tertinggi
top_5 = df.groupby('nama_kabupaten_kota')['jumlah_penderita_dm'].sum().sort_values(ascending=False).head(5)
top_5_list = [{"nama": k, "total": int(v)} for k, v in top_5.items()]

# 19. Korelasi / Insight Tambahan (Rata-rata pertumbuhan)
# Menghitung rata-rata kasus per tahun
avg_kasus_per_tahun = int(total_kasus_jabar / df['tahun'].nunique())


# ==========================================
# 3. MENYUSUN JSON FINAL
# ==========================================
data_export = {
    # Bagian 1: Inspeksi Data Awal
    "soal_1_5_overview": {
        "total_baris": int(total_baris),
        "total_kolom": int(total_kolom),
        "wilayah_unik": int(wilayah_unik),
        "total_null": total_null,
        "total_duplikat": total_duplikat,
        "status_msg": "Data valid & bersih." if total_null == 0 else "Ditemukan data kosong, telah ditangani."
    },

    # Bagian 2: Ekstrem & Filter
    "soal_6_9_ekstrem": {
        "tahun_analisis": int(tahun_terbaru),
        "tertinggi": {
            "nama": max_nama,
            "nilai": max_nilai,
            # Persentase relatif terhadap total tahun itu
            "persentase_bar": 100
        },
        "terendah": {
            "nama": min_nama,
            "nilai": min_nilai,
            "persentase_bar": int((min_nilai / max_nilai) * 100)
        }
    },

    # Bagian 3: Statistik Deskriptif
    "soal_10_14_statistik": {
        "mean": int(mean_val),
        "mean_desc": "Rata-rata kasus per wilayah",
        "median": int(median_val),
        "median_desc": "Titik tengah data",
        "std_dev": int(std_val),
        "std_desc": "Sebaran data (Simpangan Baku)",
        "q1": int(q1_val),
        "q3": int(q3_val)
    },

    # Bagian 4: Distribusi Kategori
    "soal_15_kategori": {
        "total_wilayah": int(wilayah_unik),
        "rincian": distribusi_list
    },

    # Bagian 5: Agregasi Lanjutan (16-19)
    "soal_16_19_agregasi": {
        "total_kasus_kumulatif": total_kasus_jabar,
        "rata_rata_tahunan": avg_kasus_per_tahun,
        "tren_tahunan": tren_tahun_dict, # List of {tahun: x, jumlah: y}
        "top_5_wilayah": top_5_list
    }
}

# ==========================================
# 4. SIMPAN KE FILE JSON
# ==========================================
filename_json = 'eksplorasi_full.json'
with open(filename_json, 'w') as f:
    json.dump(data_export, f, indent=4)

print(f" Sukses! File '{filename_json}' telah dibuat.")
print("   Isi file ini mencakup jawaban komputasi untuk Soal 1 sampai 19.")
print("   Silakan unduh file ini dan letakkan di folder 'data/' pada proyek Flask Anda.")
```
